```python
# Importando bibliotecas padrao
import numpy as np
import random
```

```python
# Define numero de individuos e codificacao do individuo como sendo um vetor com 8 posicoes
# em que cada posicao indica uma coluna e o numero dentro indica a linha que a rainha foi posicionada.
num_individuos = 1000
len_indv = 8
populacao_inicial = np.zeros((num_individuos, len_indv), dtype=int)
# Ja cria primeira geracao de individuos e os printa.
for i in range(num_individuos):
  for j in range(len_indv):
    populacao_inicial[i][j] = random.randint(1,len_indv)

for i in range(num_individuos):
  print(populacao_inicial[i])
```

```python
# Definicao da funcao de fitness, ela retorna o numero de rainhas que estao se atacando.
def fitness(ind):
  f=0; #f contabiliza pares de rainhas atacantes.

  for i in range(len_indv-1):
    k=1
    # Para cada par de rainhas vetor sera checado se elas se atacam.
    for j in range(i+1,len_indv):

      if(ind[i]==ind[j]): # Checa se esta na mesma linha.
        f+=1
      if(ind[i]==ind[j]+k): # Checa se esta na mesma diagonal superior.
        f+=1
      if(ind[i]==ind[j]-k): # Checa se esta na mesma diagonal inferior.
        f+=1

      k+=1

  return f

# retona_fitness atribui um valor de fitness para todos os membros da populacao passada como parametro da funcao.
def retorna_fitness(populacao):
  
  fit = []
  
  for i in range(num_individuos):
    f = fitness(populacao[i])
    fit.append(f)

  return fit
```

```python
# Metodo de selecao por torneio escolhe 3 individuos e retorna os 2 mais aptos, ou seja, com menor valor fitness.
def selecao_torneio(fitnesses):
  
  ind1 = -1
  ind2 = -1

  while ind1 == ind2:
    # Torneio 1
    sorteados = random.sample(range(0, num_individuos), 2)
    if fitnesses[sorteados[0]] < fitnesses[sorteados[1]]:
      ind1 = sorteados[0]
    else:
      ind1 = sorteados[1]

    # Torneio 2
    sorteados = random.sample(range(0, num_individuos), 2)
    if fitnesses[sorteados[0]] < fitnesses[sorteados[1]]:
      ind2 = sorteados[0]
    else:
      ind2 = sorteados[1]  
        
  return ind1,ind2    
```

```python
# Funcao para definir o cruzamento entre dois cromossomos, o cruzamento eh sempre em um ponto unico sorteado aleatoriamente.
def cruzamento(ids,populacao):
  
  ponto = random.randint(1, len_indv-1)

  pai1 = populacao[ids[0]]
  pai2 = populacao[ids[1]]

  filho1 = np.concatenate([pai1[:ponto],pai2[ponto:]])
  filho2 = np.concatenate([pai2[:ponto],pai1[ponto:]])

  return filho1,filho2
```

```python
# Seleciona sempre os individuos mais aptos, com menor fitness, para irem direto para proxima geracao.
def elitismo(fitnesses):

  id1 = fitnesses.index(min(fitnesses))

  fitnesses.pop(id1)
  id2 = fitnesses.index(min(fitnesses))

  return id1,id2
```

```python
# Realiza uma mutacao aleatoria em um dos pontos do cromossomo de acordo com uma taxada especificada nos paramentros.
def mutacao(filhos, taxa):

  for i in range(len(filhos)):
    if random.random() < taxa:
      pos = random.randint(0, len_indv-1)
      filhos[i][pos] = random.randint(1,len_indv)

  return filhos
```

```python
# Número de gerações
num_geracoes = 25

# Fitness e peso do indivíduo ótimo
fitness_otimo = 0

for it in range(num_geracoes):
  # Nova populacao vazia
  nova_populacao = np.zeros((num_individuos,len_indv), dtype=int)

  # Calcula o fitness da população
  fit = retorna_fitness(populacao_inicial)
  #print(fit)

  # Imprime o melhor individuo da geracao atual
  id_melhor = fit.index(min(fit))
  print("Solucao da geracao", it, ":",'  '.join(map(str,populacao_inicial[id_melhor])), "Fitness da solucao = ",fit[id_melhor])
  print()
  # Caso o fitness chegue a 0 quer dizer que uma solucao ja foi encontrada e o algoritmo pode parar
  if(fit[id_melhor] == fitness_otimo):
    break;
  # Elitismo
  elite = elitismo(fit.copy()) # Copiando por causa da referencia do vetor
  nova_populacao[0] = populacao_inicial[elite[0]]
  nova_populacao[1] = populacao_inicial[elite[1]]

  # Gera os filhos restantes para completar a população
  num_filhos = 2
  while num_filhos < num_individuos:

    # Seleção por torneio
    ind_vencedores = selecao_torneio(fit)
    
    # Cruzamento
    filhos = cruzamento(ind_vencedores,populacao_inicial)

    # Mutacao (50% de chance de mutação em cada indivíduo)
    filhos = mutacao(filhos,0.5)

    # Coloca os filhos na nova população
    nova_populacao[num_filhos] = filhos[0]
    nova_populacao[num_filhos+1] = filhos[1]

    # Aumenta o número de filhod
    num_filhos = num_filhos + 2

  # Substitui a populacao antiga pela atual
  populacao_inicial = nova_populacao.copy()
print("Solucao encontrada na geracao", it, ":",'  '.join(map(str,populacao_inicial[id_melhor])), "Fitness da solucao = ",fit[id_melhor])
```
